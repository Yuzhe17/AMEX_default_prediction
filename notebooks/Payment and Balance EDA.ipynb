# Payment (P) variables

```python
url_1 = "/Users/isislim/code/Yuzhe17/AMEX_default_prediction/raw_data/defaulter_data_13364.csv"
url_2 = "/Users/isislim/code/Yuzhe17/AMEX_default_prediction/raw_data/payer_data_41940.csv"
```
